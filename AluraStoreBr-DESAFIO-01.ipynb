### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
print("\n--- Faturamento Total por Loja ---")
print(f"Faturamento Loja 1: R$ {faturamento_loja:,.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:,.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

print("Dados carregados com sucesso!")
print("\nAmostra dos dados da Loja 1:")
print(loja.head())

faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print("\n--- Faturamento Total por Loja ---")
print(f"Faturamento Loja 1: R$ {faturamento_loja:,.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:,.2f}")


dados_faturamento = {
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento Total': [faturamento_loja, faturamento_loja2, faturamento_loja3, faturamento_loja4]
}
df_faturamento = pd.DataFrame(dados_faturamento).sort_values('Faturamento Total', ascending=False)

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(10, 6))

barplot = sns.barplot(
    data=df_faturamento,
    x='Loja',
    y='Faturamento Total',
    palette='viridis'
)

plt.title('Faturamento Total por Loja', fontsize=16, weight='bold')
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Faturamento Total (em Milhões de R$)', fontsize=12)

formatter = mticker.FuncFormatter(lambda x, p: f'R$ {x/1e6:.2f}M')
plt.gca().yaxis.set_major_formatter(formatter)

for index, row in df_faturamento.iterrows():
    barplot.text(index, row['Faturamento Total'], f"R$ {row['Faturamento Total']/1e6:.2f}M",
                 color='black', ha="center", va="bottom")

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

df_completo = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

print("DataFrames combinados com sucesso!")

analise_categorias = df_completo.groupby(['Loja', 'Categoria do Produto']).size().unstack(level='Loja', fill_value=0)

analise_categorias['Total Geral'] = analise_categorias.sum(axis=1)
analise_categorias = analise_categorias.sort_values('Total Geral', ascending=False).drop(columns='Total Geral')

print("\n--- Quantidade de Vendas por Categoria em Cada Loja ---")
print(analise_categorias)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 14))
fig.suptitle('Distribuição Percentual de Vendas por Categoria', fontsize=20, weight='bold')

axes = axes.flatten()
lojas = analise_categorias.columns
cores = plt.get_cmap('viridis_r')(range(len(analise_categorias)))

for i, nome_loja in enumerate(lojas):
    ax = axes[i]
    dados_loja = analise_categorias[nome_loja]

    dados_loja = dados_loja[dados_loja > 0]

    ax.pie(
        dados_loja,
        labels=dados_loja.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=cores,
        pctdistance=0.85,
        textprops={'fontsize': 10}
    )
    ax.set_title(nome_loja, fontsize=16, weight='bold')
    centre_circle = plt.Circle((0,0),0.70,fc='white')
    ax.add_artist(centre_circle)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacoes = df_completo.groupby('Loja')['Avaliação da compra'].mean().round(2)

media_avaliacoes = media_avaliacoes.sort_values(ascending=False)

print("--- Média de Avaliações por Loja (de 0 a 5) ---")
print(media_avaliacoes)
```

```python
media_avaliacoes = df_completo.groupby('Loja')['Avaliação da compra'].mean().round(2)

df_plot = media_avaliacoes.sort_values().reset_index()
df_plot.columns = ['Loja', 'Média Avaliação']

print("--- Média de Avaliações por Loja (de 0 a 5) ---")
print(df_plot)

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 7))

plt.stem(df_plot['Loja'], df_plot['Média Avaliação'], linefmt='grey', markerfmt=' ', basefmt=" ")

plt.scatter(df_plot['Loja'], df_plot['Média Avaliação'], s=400, color='royalblue', zorder=3)

for i, row in df_plot.iterrows():
    plt.text(row['Loja'], row['Média Avaliação'], f"{row['Média Avaliação']:.2f}",
             color='white', ha='center', va='center', fontsize=12, weight='bold')

plt.title('Comparativo de Satisfação do Cliente por Loja', fontsize=18, weight='bold', pad=20)
plt.ylabel('Média da Avaliação (Escala de 0 a 5)', fontsize=12)
plt.xlabel('Lojas', fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

plt.ylim(0, 5)

sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
contagem_produtos_loja = df_completo.groupby(['Loja', 'Produto']).size().reset_index(name='Quantidade Vendida')

print("--- Contagem de Vendas por Produto em Cada Loja (Amostra) ---")
print(contagem_produtos_loja.head())

top_5_mais_vendidos = contagem_produtos_loja.loc[contagem_produtos_loja.groupby('Loja')['Quantidade Vendida'].nlargest(5).index.get_level_values(1)]
top_5_menos_vendidos = contagem_produtos_loja.loc[contagem_produtos_loja.groupby('Loja')['Quantidade Vendida'].nsmallest(5).index.get_level_values(1)]

print("\n--- Top 5 Produtos MAIS Vendidos por Loja ---")
print(top_5_mais_vendidos)

print("\n--- Top 5 Produtos MENOS Vendidos por Loja ---")
print(top_5_menos_vendidos)
```

# 5. Frete Médio por Loja


```python
media_frete = df_completo.groupby('Loja')['Frete'].mean().round(2)

media_frete = media_frete.sort_values(ascending=False)

print("--- Custo Médio de Frete por Loja ---")
print(media_frete.apply(lambda x: f"R$ {x:,.2f}"))
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(10, 6))

ax = sns.barplot(
    x=media_frete.index,
    y=media_frete.values,
    palette='plasma'
)

ax.set_title('Custo Médio de Frete por Loja', fontsize=18, weight='bold')
ax.set_xlabel('Lojas', fontsize=12)
ax.set_ylabel('Custo Médio do Frete (R$)', fontsize=12)

ax.bar_label(ax.containers[0], fmt='R$ %.2f', fontsize=12)

plt.tight_layout()
plt.show()
```
